```python
from census import Census
from us import states
```

```python
import pandas as pd
```

```python
#my census api key is c7a11721ba3362e4b47283d35c4206422bba721a

c = Census("c7a11721ba3362e4b47283d35c4206422bba721a")
```

```python
#get the data for Nassau

df_tract_nassau_t = pd.DataFrame(c.acs5.state_county_tract('B03002_001E', states.NY.fips, '059', Census.ALL))
df_tract_nassau_w = pd.DataFrame(c.acs5.state_county_tract('B03002_003E', states.NY.fips, '059', Census.ALL))
df_tract_nassau_b = pd.DataFrame(c.acs5.state_county_tract('B03002_004E', states.NY.fips, '059', Census.ALL))
df_tract_nassau_h = pd.DataFrame(c.acs5.state_county_tract('B03002_012E', states.NY.fips, '059', Census.ALL))
```

```python
#merge the data for Nassau

df_tract_nassau = pd.merge(df_tract_nassau_t, df_tract_nassau_w, on=['tract','state','county'])
df_tract_nassau = pd.merge(df_tract_nassau, df_tract_nassau_b, on=['tract','state','county'])
df_tract_nassau = pd.merge(df_tract_nassau, df_tract_nassau_h, on=['tract','state','county'])
```

```python
#get the same data for suffolk

df_tract_suffolk_t = pd.DataFrame(c.acs5.state_county_tract('B03002_001E', states.NY.fips, '103', Census.ALL))
df_tract_suffolk_w = pd.DataFrame(c.acs5.state_county_tract('B03002_003E', states.NY.fips, '103', Census.ALL))
df_tract_suffolk_b = pd.DataFrame(c.acs5.state_county_tract('B03002_004E', states.NY.fips, '103', Census.ALL))
df_tract_suffolk_h = pd.DataFrame(c.acs5.state_county_tract('B03002_012E', states.NY.fips, '103', Census.ALL))
```

```python
#merge the data for suffolk

df_tract_suffolk = pd.merge(df_tract_suffolk_t, df_tract_suffolk_w, on=['tract','state','county'])
df_tract_suffolk = pd.merge(df_tract_suffolk, df_tract_suffolk_b, on=['tract','state','county'])
df_tract_suffolk = pd.merge(df_tract_suffolk, df_tract_suffolk_h, on=['tract','state','county'])
```

```python
#union the two dataframes from long island

frames = [df_tract_nassau,df_tract_suffolk]

df_tract_longisland = pd.concat(frames)
```

```python
df_tract_longisland = df_tract_longisland.rename(columns={"B03002_001E":"pop_total",\
                                                             "B03002_003E":"pop_white",\
                                                             "B03002_004E":"pop_black",\
                                                             "B03002_012E":"pop_hispanic"})
```

```python
df_tract_longisland = df_tract_longisland.filter(items=['tract',
                                                         'state', 
                                                         'county',
                                                         'pop_total',
                                                         'pop_white',
                                                         'pop_black',
                                                         'pop_hispanic'])
```

```python
#write this data to a csv file

df_tract_longisland.to_csv('long_island_tract_demographics.csv',index = False, header=True)
```
