```python
import pyzillow
from pyzillow.pyzillow import ZillowWrapper, GetDeepSearchResults, GetUpdatedPropertyDetails
import pandas as pd
import xlrd
```

```python
df_sold_homes = pd.read_excel("CurrentSearchResults.xls")
```

```python
#df_sold_homes['Street Address'].iloc[0]
df_sold_homes.columns
```

```python
zillow_data = ZillowWrapper('X1-ZWz183vfyfl8gb_4srxq')
```

```python
#create the data set
comp_data = {#from GetDetailedSearchResults
             'address':[],
             'home_size':[],
             'property_size':[],
             'home_type':[],
             'last_sold_date':[],
             'last_sold_price':[],
             'tax_value':[],
             'tax_year':[],
             #from GetUpdatedPropertyDetails
             'bedrooms':[],
             'bathrooms':[],
             'agent_name':[],
             'neighborhood':[],
             'year_built':[]
            }

df_comp_data = pd.DataFrame(comp_data)

#run through a list of property addresses and zip codes for info

for listing in range(0,len(df_sold_homes)):
    #you need to pull the address and the zip code from each record of the data frame
    address = df_sold_homes['Street Address'].iloc[listing]
    zip_code = df_sold_homes['Zip'].iloc[listing]
    
    #this is where you get the deep search results for each house, 1 at a time
    deep_search_response = zillow_data.get_deep_search_results(address,zip_code, True)
    deep_search_result = GetDeepSearchResults(deep_search_response)
    
    #this is where you get the updated property details.  This requires the zillow_id from deep search results.
    try:
        zillow_id = deep_search_result.zillow_id
        updated_property_details_response = zillow_data.get_updated_property_details(zillow_id)
        updated_property_result = GetUpdatedPropertyDetails(updated_property_details_response)
    except:
        pass
    
    #pull variable from deep search results
    home_size = updated_property_result.home_size
    property_size = deep_search_result.property_size
    home_type = deep_search_result.home_type
    last_sold_date = deep_search_result.last_sold_date
    last_sold_price = deep_search_result.last_sold_price
    tax_value = deep_search_result.tax_value
    tax_year = deep_search_result.tax_year
    
    #pull variables from updated property results
    bedrooms = updated_property_result.bedrooms
    bathrooms = updated_property_result.bathrooms
    agent_name = updated_property_result.agent_name
    neighborhood = updated_property_result.neighborhood
    year_built = updated_property_result.year_built
    
    #append all this data to the dataframe
    df_comp_data = df_comp_data.append({#from GetDetailedSearchResults
                                         'address':address,
                                         'home_size':home_size,
                                         'property_size':property_size,
                                         'home_type':home_type,
                                         'last_sold_date':last_sold_date,
                                         'last_sold_price':last_sold_price,
                                         'tax_value':tax_value,
                                         'tax_year':tax_year,
                                         #from GetUpdatedPropertyDetails
                                         'bedrooms':bedrooms,
                                         'bathrooms':bathrooms,
                                         'agent_name':agent_name,
                                         'neighborhood':neighborhood,
                                         'year_built':year_built
                                        }, ignore_index=True
                                       )
    
    
    #print the results here  
    print(df_comp_data.loc[listing])
```
