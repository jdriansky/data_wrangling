```python
import pandas as pd
```

```python
#import the csv file as a data frame
daily_stock = pd.read_csv('/Users/jeremiahdriansky/Downloads/AAPL.csv')
```

```python
#output the data frame
daily_stock
```

```python
#substring the date to get a month field
year_month = daily_stock.Date.str.slice(0,7)
```

```python
#add the month field to the data frame
daily_stock['year_month'] = year_month
```

```python
#group the opening stock price by the substringed date field
monthly_average = daily_stock.groupby(year_month).Open.mean()
```

```python
#plot the monthly averages
monthly_average.plot(kind='bar',title='monthly aapl stock price')
```
